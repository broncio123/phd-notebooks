```python
import matplotlib.pyplot as plt
import numpy
%matplotlib notebook
```

```python
Sequences = {
    'cWza' : 'APLVRWNRVISQLVPTITGVHDLTETVRYIKT',
    'cWza-K375C' : 'APLVRWNRVISQLVPTITGVHDLTETVRYICT',
    'cWza-S355C' : 'APLVRWNRVICQLVPTITGVHDLTETVRYIKT',
    'cWza-Y373C' : 'APLVRWNRVISQLVPTITGVHDLTETVRCIKT'
}

mutant_names = ['cWza','cWza-K375C','cWza-S355C','cWza-Y373C']
mutant_names.reverse()
```

```python
fig,ax = plt.subplots(1,1,dpi=200)
fig.patch.set_visible(False)
ax.axis('off')

Y0 = 1
for mutant in mutant_names:
    seq = Sequences[mutant]

    scale = 12
    X = numpy.linspace(0,scale*len(seq),len(seq))
    Y = Y0*numpy.ones(len(seq))

    fs0 = 10
    off_set_x = -100
    
    for i in range(len(seq)):
        if str(seq[i]) == 'C':
            res_color = 'magenta'
            ax.text(X[i],Y[i],str(seq[i]),color=res_color,fontsize=fs0,fontweight='bold')
        else:
            res_color = 'black'
            ax.text(X[i],Y[i],str(seq[i]),color=res_color,fontsize=fs0)
    ax.text(off_set_x,Y0,mutant,fontsize=fs0,fontstyle='italic')
    Y0 += 0.1

fs1 = 10
resn0 = 345
off_set_y = 0.1
resn = resn0 + numpy.asarray(range(len(seq)))    
for i in range(len(seq)):
    ax.text(X[i],Y0+off_set_y,str(resn[i]),rotation=90,fontsize=fs1)
    ax.text(X[i],Y0-5.5*off_set_y,str(i+1),rotation=90,fontsize=fs1)

ax.text(off_set_x,Y0+off_set_y,'native resn',fontsize=fs1)    
ax.text(off_set_x,Y0-5.5*off_set_y,'docked resn',fontsize=fs1)  

ax.set_xlim(off_set_x,X[-1])
ax.set_ylim(off_set_y,Y0+off_set_y)
fig.tight_layout()
plt.show()
```
